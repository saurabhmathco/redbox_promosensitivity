# Week vs Transactionid

```python
import datetime
start_date = pd.to_datetime("2018-02-01").date()
end_date = pd.to_datetime("2018-05-01").date()
day_delta=datetime.timedelta(days=1)
count=0

for i in range(0,(end_date - start_date).days,7):
    start_i=(start_date+i*day_delta)
    end_i=(start_date+(i+7)*day_delta)
    x=trans4_df['new_date_column'].loc[(trans4_df['new_date_column'] >= start_i) & (trans4_df['new_date_column'] < end_i)]
    y=trans4_df.count_transactionid.iloc[i:i+7]

    count+=1
    
    figure,axes=plt.subplots(nrows=1,ncols=1,figsize=(12,6))
    axes.plot(x,y,color="darkblue",lw=3,ls="-")
    plt.legend(["Week"+" "+str(count)])
```
